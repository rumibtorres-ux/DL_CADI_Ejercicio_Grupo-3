```python
import numpy as np

def step(z):
    return 1 if z >= 0 else 0

def neurona(x1, x2, w1, w2, b):
    z = x1*w1 + x2*w2 + b
    y = step(z)
    return z, y

# Valores base
w1, w2, b = 1, 1, -2.0   # intenta parecerse a una compuerta AND

casos = [(0,0),(0,1),(1,0),(1,1)]
for x1, x2 in casos:
    z, y = neurona(x1, x2, w1, w2, b)
    print(f"Entrada:({x1},{x2})  z={z:.1f}  salida={y}")


```

**Preguntas**

1. ¿Con qué valor de b la neurona se comporta más parecido a AND?

Con b = -1.5 porque solo se activa cuando las dos entradas son 1

2. ¿Qué efecto tiene bajar o subir el bias en la salida?

El bias controla qué tan fácil se activa la neurona.
Si el bias es menos negativo se activa más fácil, si es más negativo se vuelve más estricta.

menos negativo → más fácil activarse
más negativo → más difícil activarse
