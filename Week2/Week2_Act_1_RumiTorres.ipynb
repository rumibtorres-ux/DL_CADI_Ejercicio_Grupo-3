**Datos y preprocesamiento**

```python

# Datos y preprocesamiento


import numpy as np

# Dataset simple (problema tipo AND)
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y = np.array([0, 0, 0, 1])  # salida esperada AND
```

**Baseline Neurona / Perceptrón simple**

```python
# Entrenar baseline y reportar métricas

def step(z):
    return 1 if z >= 0 else 0

def perceptron(X, w, b):
    z = np.dot(X, w) + b
    y_pred = np.array([step(val) for val in z])
    return z, y_pred

# Pesos y bias iniciales
w = np.array([1, 1])
b = -1.5

z, y_pred = perceptron(X, w, b)

print("z:", z)
print("Predicciones:", y_pred)
print("Reales:", y)
```

**Métricas y matriz de confusión**

```python
from sklearn.metrics import accuracy_score, confusion_matrix

accuracy = accuracy_score(y, y_pred)
matrix = confusion_matrix(y, y_pred)

print("Accuracy:", accuracy)
print("Matriz de confusión:")
matrix
```

**MPL** (extensión conceptual del perceptrón)

```python


# Entrenar MLP y reportar métricas

# Nuevos parámetros
w_mlp = np.array([1, 1])
b_mlp = -0.5

z_mlp, y_pred_mlp = perceptron(X, w_mlp, b_mlp)

print("Predicciones MLP:", y_pred_mlp)
print("Accuracy MLP:", accuracy_score(y, y_pred_mlp))
print("Matriz de confusión MLP:")
confusion_matrix(y, y_pred_mlp)
```

¿Qué efecto tiene bajar o subir el bias en la salida?
El bias controla qué tan fácil se activa la neurona. Si el bias es menos negativo se activa más fácil, si es más negativo se vuelve más estricta.

menos negativo → más fácil activarse más negativo → más difícil activarse


# **Conclusiones**

* El sesgo (bias) define el umbral de activación de la neurona y determina qué
tan fácil o difícil es que esta se active. En los ejercicios realizados se pudo observar que, cuando el bias es menos negativo (por ejemplo, b = -0.5), la neurona se activa con mayor facilidad, generando salidas iguales a 1 incluso cuando solo una de las entradas tiene valor 1. Por el contrario, cuando el bias toma valores más negativos, como b = -1.5 o b = -2.0, la neurona necesita una suma ponderada mayor para poder activarse.

* Al mantener los pesos constantes, se evidenció que pequeñas variaciones en el bias pueden generar cambios importantes en la clasificación. En todos los casos se mantuvieron los valores w1 = 1 y w2 = 1, y aun así los resultados cambiaron al modificar únicamente el bias. Esto ocurre porque el bias altera el valor de z, lo que finalmente afecta la salida de la neurona. Por esta razón, se puede afirmar que el bias es un parámetro fundamental en la toma de decisiones de una neurona artificial, incluso cuando los pesos permanecen iguales.

* Finalmente, se observó que el perceptrón puede comportarse como una compuerta lógica AND cuando el bias está correctamente configurado. En el caso de b = -1.5, la neurona solo se activa cuando ambas entradas son iguales a 1, lo cual coincide con el funcionamiento de la compuerta lógica AND. Esto demuestra que una neurona simple es capaz de resolver problemas linealmente separables mediante un ajuste adecuado de los pesos y del sesgo.
